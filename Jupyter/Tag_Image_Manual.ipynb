```python
# Jobs-31: HSV และ RGB Color scale, Histogram, Split Channel

import cv2
import numpy as np
import matplotlib.pyplot as plt

# โหลดภาพ
img = cv2.imread('./image/HighWay.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# ฟังก์ชันคลิกแสดงค่า pixel
def click_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        rgb_val = img_rgb[y, x]
        hsv_val = img_hsv[y, x]
        print(f"RGB @({x},{y}) = {rgb_val}")
        print(f"HSV @({x},{y}) = {hsv_val}")
        font = cv2.FONT_HERSHEY_SIMPLEX
        txt = f"{x},{y}"
        cv2.putText(img_copy, txt, (x, y), font, 0.5, (255, 255, 255), 1)
        cv2.imshow("Click Image", img_copy)

img_copy = img_rgb.copy()
cv2.imshow("Click Image", img_copy)
cv2.setMouseCallback("Click Image", click_event)

# แสดงภาพ RGB และ HSV
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
plt.title("HSV Image")
plt.axis('off')
plt.show()

# แสดง Histogram RGB
color = ('r','g','b')
plt.figure()
for i, col in enumerate(color):
    hist = cv2.calcHist([img_rgb],[i],None,[256],[0,256])
    plt.plot(hist, color = col)
plt.title("RGB Histogram")
plt.show()

# แสดง Histogram HSV
h,s,v = cv2.split(img_hsv)
plt.figure()
plt.plot(cv2.calcHist([h],[0],None,[256],[0,256]), label='H', color='m')
plt.plot(cv2.calcHist([s],[0],None,[256],[0,256]), label='S', color='c')
plt.plot(cv2.calcHist([v],[0],None,[256],[0,256]), label='V', color='y')
plt.legend()
plt.title("HSV Histogram")
plt.show()

# แสดง split RGB
r,g,b = cv2.split(img_rgb)
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.imshow(r, cmap='Reds')
plt.title("Red Channel")
plt.subplot(1,3,2)
plt.imshow(g, cmap='Greens')
plt.title("Green Channel")
plt.subplot(1,3,3)
plt.imshow(b, cmap='Blues')
plt.title("Blue Channel")
plt.show()

# แสดง split HSV
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.imshow(h, cmap='hsv')
plt.title("Hue Channel")
plt.subplot(1,3,2)
plt.imshow(s, cmap='gray')
plt.title("Saturation Channel")
plt.subplot(1,3,3)
plt.imshow(v, cmap='gray')
plt.title("Value Channel")
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Step 31 - HSV VS GRB Yellow Mango
# https://www.linuxtut.com/en/ff4a07d73e14385b6922/

import cv2 as cv
import numpy
from matplotlib import pyplot as plt

frame = numpy.ones([256, 1], dtype=int)
frame[0:4] = 0
frame[252:256] = 0

def show_img(imgRGB):
    imgHSV = cv.cvtColor(imgRGB, cv.COLOR_BGR2HSV)
    b, g, r = imgRGB[:, :, 0], imgRGB[:, :, 1], imgRGB[:, :, 2]

    hist_b = cv.calcHist([b], [0], None, [256], [0, 256]) * frame
    hist_g = cv.calcHist([g], [0], None, [256], [0, 256]) * frame
    hist_r = cv.calcHist([r], [0], None, [256], [0, 256]) * frame

    plt.plot(hist_r, color='r', label="r")
    plt.plot(hist_g, color='g', label="g")
    plt.plot(hist_b, color='b', label="b")
    plt.legend()
    plt.show()

    h, s, v = imgHSV[:, :, 0], imgHSV[:, :, 1], imgHSV[:, :, 2]
    hist_h = cv.calcHist([h], [0], None, [256], [0, 256]) * frame
    hist_s = cv.calcHist([s], [0], None, [256], [0, 256]) * frame
    hist_v = cv.calcHist([v], [0], None, [256], [0, 256]) * frame

    plt.plot(hist_h, color='r', label="h")
    plt.plot(hist_s, color='g', label="s")
    plt.plot(hist_v, color='b', label="v")
    plt.legend()
    plt.show()

    return hist_r, hist_g, hist_b, hist_h, hist_s, hist_v

img = cv.imread('./image/ME.jpg', 1)  # 1 = Color
#img = cv.imread('./image/Chili_Grn.jpg', 1)  # 1 = Color
#img = cv.imread('./image/Chili_Ylw.jpg', 1)  # 1 = Color

r, g, b, h, s, v = show_img(img)

cv.imshow("ธนวัฒน์ วัฒนกิจจา", img)
cv.waitKey(0)  # Click ที่รูปคลิกใดๆ จะปิดรูปภาพ
cv.destroyAllWindows()
```

```python
import cv2
import numpy as np
from PIL import Image

# โหลดลายน้ำ
pil_logo = Image.open('C:/Users/BiscuitRulez/OneDrive/Desktop/Wk03_OpenCV(2) and Video/image/SUT.png')
pil_logo.thumbnail((250, 250))

# เปิดกล้อง
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # กล้องเปล่า
    cv2.imshow('Camera (No Logo)', frame)

    # กล้องพร้อมลายน้ำ
    color_converted = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pil_image = Image.fromarray(color_converted)
    pil_image.paste(pil_logo, (0, 0), pil_logo)
    watermarked = np.array(pil_image)
    watermarked_bgr = cv2.cvtColor(watermarked, cv2.COLOR_RGB2BGR)
    cv2.imshow('Camera with Logo', watermarked_bgr)

    # ออกจาก loop เมื่อกด 'q'
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
